```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
test_file = pd.read_csv("test.csv")
test_file.head(3)
```

## Har bir platforma foydalanuvchilari soni va eng ko'p odamlar qaysi platforma ishlatishadi ?

```python
platform_counts = test_file['Platform'].value_counts()
plt.figure(figsize=(10, 6))
plt.bar(platform_counts.index, platform_counts.values, color='b')
plt.title('Platforma foydalanuvchilari', fontsize=16)
plt.xlabel('Ijtimoiy tarmoqlar', fontsize=14)
plt.ylabel('Foydalanuvchilar soni', fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', linewidth=0.7)
plt.show()
```

### Korishimiz mumkin eng ko'p foydalanuvchilar Facebookda ekan eng kami esa Whatsapp ekan haqiqatga yaqin menimcha men ham ishlatmayman Whatsapp :)


## Ijtimoiy tarmoqlardan erkaklar ko'proq foydalanadimi yoki ayollar ?


```python
gender_counts= test_file["Gender"].value_counts()
plt.bar(gender_counts.head(3).index, gender_counts.head(3).values)
plt.title("Gender bo'yicha foydalanuvchilar soni", fontsize=14)
plt.xlabel("Types of Gender", fontsize=12)
plt.ylabel("Foydalanuvchilar soni", fontsize=13)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', linewidth=0.4)
plt.show()
```

### Ko'rishimiz mumkin ijtimoiy tarmoqlardan ko'proq Erkaklar foydalanisharkan.

## Qaysi ijtimoiy tarmoqga eng ko'p post yuklanadi ?


```python
post_to_platform=test_file.groupby("Platform")["Posts_Per_Day"].sum()

plt.bar(post_to_platform.index, post_to_platform.values, color="yellow")
plt.xticks(rotation=45, fontsize=12)
plt.title("Ijtimoiy tarmoqlar kunlik postlar", fontsize=14)
plt.xlabel("Ijtimoiy tarmoqlar", fontsize=13)
plt.ylabel("Foydalanuvchilar soni", fontsize=13)
plt.show()
```

## Qaysi tarmoqda ko'proq LIKElar soni ?

```python
likes_per_day=test_file.groupby("Platform")["Likes_Received_Per_Day"].sum()

plt.bar(likes_per_day.index, likes_per_day.values, color="red")
plt.xticks(rotation=45, fontsize=12)
plt.title("Ijtimoiy tarmoqlarda kunlik LIKElar", fontsize=14)
plt.xlabel("Ijtimoiy tarmoqlar", fontsize=13)
plt.ylabel("Foydalanuvchilar soni", fontsize=13)
plt.show()
```

## Erkaklar va ayollar soni ?

```python
num_women=test_file[test_file["Gender"]=="Female"]["User_ID"].count()
num_men=test_file[test_file["Gender"]=="Male"]["User_ID"].count()
fig, ax = plt.subplots()
ax.bar("Women", num_women)
ax.bar("Men", num_men)
ax.set_title("Ayollar va erkaklar soni")
ax.set_ylabel("Count")
plt.show()
```

### Platformalar kesimida ortacha postlar ?


```python
posts_stats = test_file.groupby('Platform')['Posts_Per_Day'].mean().reset_index()
plt.figure(figsize=(10, 6))
plt.bar(posts_stats['Platform'], posts_stats['Posts_Per_Day'])
plt.title('Ortacha kunlik postlar platformalarda ')
plt.xlabel('Platformalar')
plt.ylabel('Ortacha postlar')
plt.show()
```
